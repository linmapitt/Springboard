```python
# Find businessparking attribute entries

bus_df_attr = bus_df['attributes']
type(bus_df_attr[0])
len = bus_df_attr.shape[0]
parking_df = pd.DataFrame(columns=['business_id', 'stars','garage','street','validated','lot','valet'])
i = 0
j=0
while i < len-1:
    if type(bus_df_attr[i]) == str:
        X1 = bus_df_attr[i].replace("u'", "'")
        X2 = X1.replace("'{", "{")
        X3 = X2.replace("}'", "}")
        X4 = X3.replace("''","'")
        tmp_dict = eval(X4)
        for key, value in tmp_dict.items():
            if key =='BusinessParking' and type(value)==dict:
                if value.has_key('garage'):
                    garage_str = value['garage']
                else:
                    garage_str = False
                    
                if value.has_key('street'):
                    street_str = value['street']
                else:
                    street_str = False
                    
                if value.has_key('validated'):
                    valid_str = value['validated']
                else:
                    valid_str = False
                    
                if value.has_key('lot'):
                    lot_str = value['lot']
                else:
                    lot_str = False
                    
                if value.has_key('valet'):
                    valet_str = value['valet']
                else:
                    valet_str = False
                    
                parking_df.loc[j] = [bus_df['business_id'][j], bus_df['stars'][j], garage_str, street_str, valid_str, lot_str, valet_str]
                j = j+1

    i = i+1


parking_df.head()
```

```python
parking_df['stars'].value_counts()
```

```python
# It looks like missing values in parking_df is <10. 
# I replaced them with majority value (False in this case)
# Furthermore, I need to integrate stars into two big categories (<3 will be 0, >=3 will be 1)

parking_df['high_rate'] = parking_df['stars'] >= 3
```

```python
parking_df.head()
```

```python
# Use feature selection to pick the most important feature for business parking
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
X = parking_df[['garage','street','validated','lot','valet']]
Y = parking_df['high_rate']
```

```python
test = SelectKBest(score_func=chi2, k=1)
fit = test.fit(X, Y)
print(fit.scores_)
```

```python
Selected_feature_names = X.columns[test.get_support()]
Selected_feature_names
```

```python
# It looks like street parking plays the most important role in business parking regarding rate
# I will use street parking feature in the final dataframe.
```

```python
len = bus_df_1.shape[0]
i = 0
my_disc = {}
my_df = pd.DataFrame(columns=['business_id', 'stars', 'takeout_attr','price_attr','group_attr','kids_attr','resv_attr',
                              'deliv_attr','attire_attr','outdoor_attr','bike_attr','creditcard_attr','WiFi_attr'])

while i < len-1:
    tmp_str = ""
    tmp_str = bus_df_1['attributes'][i]
    if type(tmp_str) == str:
        X1 = tmp_str.replace("u'", "'")
        X2 = X1.replace("'{", "{")
        X3 = X2.replace("}'", "}")
        X4 = X3.replace("''","'")
        tmp_dict = eval(X4)
        
        if tmp_dict.has_key('RestaurantsTakeOut'):
            Takeout_attr = tmp_dict['RestaurantsTakeOut']
        else:
            Takeout_attr = True
         
        if tmp_dict.has_key('RestaurantsPriceRange2'):
            if tmp_dict['RestaurantsPriceRange2'] =='None':
                Price_attr = 2
            else:
                Price_attr = tmp_dict['RestaurantsPriceRange2']
        else:
            Price_attr = 2

        if tmp_dict.has_key('RestaurantsGoodForGroups'):
            Group_attr = tmp_dict['RestaurantsGoodForGroups']
        else:
            Group_attr = True

        if tmp_dict.has_key('GoodForKids'):
            Kids_attr = tmp_dict['GoodForKids']
        else:
            Kids_attr = True
        
        if tmp_dict.has_key('RestaurantsReservations'):
            Resv_attr = tmp_dict['RestaurantsReservations']
        else:
            Resv_attr = False  

        if tmp_dict.has_key('RestaurantsDelivery'):
            Deliv_attr = tmp_dict['RestaurantsDelivery']
        else:
            Deliv_attr = False

        if tmp_dict.has_key('RestaurantsAttire'):
            Attire_attr = tmp_dict['RestaurantsAttire']
        else:
            Attire_attr = 'casual'
            
        if tmp_dict.has_key('OutdoorSeating'):
            Outdoor_attr = tmp_dict['OutdoorSeating']
        else:
            Outdoor_attr = False   
        
        if tmp_dict.has_key('BikeParking'):
            Bike_attr = tmp_dict['BikeParking']
        else:
            Bike_attr = True
        
        if tmp_dict.has_key('BusinessAcceptsCreditCards'):
            Creditcard_attr = tmp_dict['BusinessAcceptsCreditCards']
        else:
            Creditcard_attr = True 
            
       
        my_df.loc[i] = [bus_df_1['business_id'][i], bus_df_1['stars'][i], Takeout_attr, Price_attr,Group_attr,Kids_attr,
                        Resv_attr, Deliv_attr, Attire_attr, Outdoor_attr, Bike_attr, Creditcard_attr]
    i = i + 1
    #if i==15:
        #break;
        
my_df
```
